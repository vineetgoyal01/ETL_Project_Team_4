```python
#Define path to .csv file
abs_csv_path = "ABS_Table2.csv"

# Import the data into a Pandas DataFrame
abs_df = pd.read_csv(abs_csv_path)
abs_df.head()
```

```python
#Filtering the year
abs_lgas_df = abs_df.loc[abs_df["Year"] == 2016.0, :]

#Filtering the LGAs 
abs_melb_lgas = abs_lgas_df.loc[(abs_lgas_df["Label"] == "Melbourne (C)") | 
                                (abs_lgas_df["Label"] == "Port Phillip (C)") | 
                                (abs_lgas_df["Label"] == "Stonnington (C)") |
                                (abs_lgas_df["Label"] == "Yarra (C)") | 
                                (abs_lgas_df["Label"] == "Banyule (C)") |
                                (abs_lgas_df["Label"] == "Bayside (C)") |
                                (abs_lgas_df["Label"] == "Boroondara (C)") | 
                                (abs_lgas_df["Label"] == "Darebin (C)") |
                                (abs_lgas_df["Label"] == "Glen Eira (C)") |
                                (abs_lgas_df["Label"] == "Hobsons Bay (C)") | 
                                (abs_lgas_df["Label"] == "Kingston (C) (Vic.)") | 
                                (abs_lgas_df["Label"] == "Manningham (C)") |
                                (abs_lgas_df["Label"] == "Maribyrnong (C)") |
                                (abs_lgas_df["Label"] == "Monash (C)") |
                                (abs_lgas_df["Label"] == "Moonee Valley (C)") |
                                (abs_lgas_df["Label"] == "Moreland (C)") | 
                                (abs_lgas_df["Label"] == "Whitehorse (C)") | 
                                (abs_lgas_df["Label"] == "Brimbank (C)") | 
                                (abs_lgas_df["Label"] == "Cardinia (S)") |
                                (abs_lgas_df["Label"] == "Casey (C)") | 
                                (abs_lgas_df["Label"] == "Frankston (C)") | 
                                (abs_lgas_df["Label"] == "Greater Dandenong (C)") | 
                                (abs_lgas_df["Label"] == "Hume (C)") | 
                                (abs_lgas_df["Label"] == "Knox (C)") | 
                                (abs_lgas_df["Label"] == "Maroondah (C)") | 
                                (abs_lgas_df["Label"] == "Melton (C)") |  
                                (abs_lgas_df["Label"] == "Mornington Peninsula (S)") |
                                (abs_lgas_df["Label"] == "Nillumbik (S)") |
                                (abs_lgas_df["Label"] =="Whittlesea (C)") |
                                (abs_lgas_df["Label"] =="Wyndham (C)") |
                                (abs_lgas_df["Label"] =="Yarra Ranges (S)"), :]
abs_melb_lgas.head()
```

```python
#Set index to LGAs
#abs_clean_lgas = alb_melb_lgas.set_index("Label")

#Replace column name "Label" to "LGAs"
abs_melb_lgas = abs_melb_lgas.rename(columns=
    {"Label": "LGAs"})
abs_melb_lgas.head()

#Remove text within parentheses
abs_melb_lgas["LGAs"] = abs_melb_lgas["LGAs"].str.replace(r"(\(.+\))","")
#Remove remaining whitespace from LGA name
abs_melb_lgas["LGAs"] = abs_melb_lgas["LGAs"].str.rstrip()

abs_melb_lgas
```

```python
#Select columns to match the other DataFrame by Vineet
abs_lgas_clean_df = pd.DataFrame()
abs_lgas_clean_df["S.No"] = abs_melb_lgas["Code"]
abs_lgas_clean_df["LGAs"] = abs_melb_lgas["LGAs"]
abs_lgas_clean_df["Year"] = abs_melb_lgas["Year"]
abs_lgas_clean_df["Adult_Population"] = abs_melb_lgas["Total population aged 15 years and over (no.)"]
#abs_clean_df.set_index("S.No", inplace=True)
abs_lgas_clean_df = abs_lgas_clean_df.rename(columns=
    {"LGAs": "Local Government Area"})
abs_lgas_clean_df.head()
```

```python
abs_lgas_clean_df.to_csv("ABS LGas Clean.csv", index=False, header=True)
```

```python
# Import of Suburb list with price (exported from Jupyter Notebook 1)

suburb_list_df = pd.read_csv('LGA_final_summary.csv', low_memory=False)

suburb_list_df.head()
```

```python
# Merging of LGA data

lga_data_df = pd.merge(abs_lgas_clean_df, suburb_list_df, on="Local Government Area", how="right")
lga_data_df.head()
```

```python
# Drop null rows
final_lga_data_df = lga_data_df.dropna()
final_lga_data_df.head()
```

```python
# Write .csv
final_lga_data_df.to_csv("final_lga_demographic.csv")
```
