```python
#Web Scraping 1- need to interact with a tbale on Wikipedia 
#add in the local coucil area data in order to aggregate 
#webscraping in order to figure out the list of Local Government LGAs (from Wikipedia)

lgas_melbourne_url ="https://en.wikipedia.org/wiki/Local_government_areas_of_Victoria"
tables = pd.read_html(lgas_melbourne_url )
tables



lga_df=tables[0]
lga_df=lga_df["Local government area"]
lga_df

#This web-scraped table gave the Names of all Local Councils within Melbourne
```

Transforming 1- need to transform this data in order to interact it with the COuncil website (another web-scraping website)
#Need to change the City of/ Shire of prefix to a City/Shire suffix

```python
#Transforming the City of
city_of=lga_df[lga_df['Local government area'].str.contains("City")]

#now delete the "City of part"
city_of['Local government area'] = city_of['Local government area'].str[8:]
city_of['Local government area'] = city_of['Local government area'].astype(str) + ' City' 

city_of
```

```python
#changing the prefix of Shire of to a Shire suffix for all Shires
shire_of=lga_df[lga_df['Local government area'].str.contains("Shire")]
shire_of['Local government area'] = shire_of['Local government area'].str[9:]
shire_of['Local government area'] = shire_of['Local government area'].astype(str) + ' Shire' 

shire_of
```

```python
#now put the datasets together 
frames=[city_of, shire_of]
lga_final_list = pd.concat(frames)
lga_final_list=lga_final_list.sort_index()
lga_final_list
lga_values=lga_final_list['Local government area'].tolist()
```

```python
#attempting to do the loop in order to get the suburbs:
#webscraping the website
suburb_sample=[]
for suburb in lga_values:
    local_council_url ="https://knowyourcouncil.vic.gov.au/councils"
# Retrieve page with the requests module
    browser.visit(local_council_url)
#Clicking on each individual council website to get the suburbs
    browser.click_link_by_partial_text(suburb)
    local_council_html=browser.html 
#Finding a list of the suburbs using beautiful soup
    local_council_soup=bs(local_council_html, 'html.parser')
    local_council = local_council_soup.find('div', class_='col-xs-12 col-sm-6 localities')
#finding paragraph of the suburb list
    suburb_text=local_council.find("p").text

#editing some of the paragraphs in order to remove ands and foolstops and other formatting issues 
    
    #making spaces in the suburb text easier to deal with (if this transformation was not applied, the and within suburb names gets converted)
    suburb_text=suburb_text.replace(" ", "_")
    #replacing and at the end of the paragraph 
    suburb_text=suburb_text.replace("_and", ",")
    #removing the underscores back to space
    suburb_text=suburb_text.replace("_", " ")
    #removing the fullstop at the end of the paragraph
    suburb_text=suburb_text.replace(".", "")
    #some text randomly had these characters to this had to be removed as well
    suburb_text = suburb_text.replace('\xa0','')
    suburb_text

    #converting the text paragraph to a list
    my_list = suburb_text.split(", ")

#Making a dictionary with the key as suburbs name as values as the list of suburbs
    greater_melbourne={
        suburb:my_list
    }

#appending this dictionary to a list 
    
    suburb_sample.append(greater_melbourne)

#time sleeping to not overload the browser
    time.sleep(2)
    
browser.quit()
```

```python
#converting the list of dictionaries to a dataframe 
suburbs_df=pd.DataFrame(suburb_sample).melt()
suburbs_df.dropna()
```

```python
#Converting the first data srouce list 
suburbs_df=suburbs_df.explode('value')
#cleaning by dropping the null values. It was noticed that the explode feature created a lot of extra rows so feel fins about dropping them 
suburbs_df=suburbs_df.dropna()
suburbs_df
```

```python
#getting rid of space before the suburb name by converting suburb name to list
suburb_greater_list=suburbs_df['value'].to_list()
```

```python
#code for removing spaces through iterating through list

suburb_space_removed=[]
for i in suburb_greater_list:
    i=i.lstrip()
    suburb_space_removed.append(i)

```

```python
suburbs_df["Suburb Name"]=suburb_space_removed
#renaming columns
suburbs_df.rename(columns={'variable':'Local Government Area'}, inplace=True)
#resetting index
suburbs_df=suburbs_df.reset_index()
#renaming columns
suburbs_df=suburbs_df[["Local Government Area", "Suburb Name"]]

```

```python
#Interacting the web-scraped data-frame with the Domain API
#creating a new column for the Suburbs ID. This ID means it can interact with other Domain APIs
#Postcode of suburs also collected from this API 
suburbs_df['ID'] = ""
suburbs_df['Postcode']=""
```

```python
for index, row in suburbs_df.iterrows():
    
    suburb=row["Suburb Name"]
    
    try:
        url = ("https://api.domain.com.au/v1/addressLocators?searchLevel=Suburb&suburb={0}&state=VIC").format(suburb)
        auth = {"X-Api-key":Api_key4}
        request = requests.get(url,headers=auth)
        r=request.json()
    
        suburbs_df.loc[index, 'ID'] = r[0]['ids'][0]['id']
        suburbs_df.loc[index, "Postcode"]=r[0]["addressComponents"][1]['shortName']
    
    #in the case of a JSONDecode Error, this ensures the code will keep running
    except (KeyError, IndexError, ValueError):
        suburbs_df.loc[index, 'ID'] = "No ID"

```

```python

final_suburb1['2016 Median Sold Price'] = ""
final_suburb2['2016 Median Sold Price'] = ""
```

```python
for index, row in final_suburb1.iterrows():
    
    id=row["ID"]
    
    try:
        url = ("https://api.domain.com.au/v1/suburbPerformanceStatistics?state=VIC&suburbId={0}&propertyCategory=house&chronologicalSpan=12&tPlusFrom=5&tPlusTo=5&values=MedianSoldPrice").format(id)
        auth = {"X-Api-key":Api_key3}
        request = requests.get(url,headers=auth)
        r=request.json()
        
        #obtaining the values needed from the API - Jan 2020 Median, Apr 2020 Median, Jan 2020 Number sold and Apr 2020 number sold
        final_suburb1.loc[index, '2016 Median Sold Price'] = r['series']['seriesInfo'][0]['values']['medianSoldPrice']
    #allows the code to keep running 
    except (KeyError, NameError, IndexError, ValueError):
        final_suburb1.loc[index, '2016 Median Sold Price'] = 'none'

```
